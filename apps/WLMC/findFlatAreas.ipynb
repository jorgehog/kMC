```python
N = 500
noise = 0.1
n = 3

signal = ones(N)

dN = N/(2*n)

phase = 0
f = 0.5
mag = 1

signal*=mag

for i in range(n):

    roughness = mag*((-1)**i*sin(linspace(0, 2*pi/f, dN) + phase) + 1)
    signal[2*i*dN:(2*i+1)*dN] = roughness
    
    
signal *= (1 + noise*random.uniform(size=N))
signal = signal - signal.min()
    
plot(signal)
show()
```

```python
def estimateFlatness(signal):
    return signal.min()/signal.mean()

print estimateFlatness(signal), estimateFlatness(signal[9*N/10:])
```

```python
min_window = N/10
increment_size = 1
flatness_critereon = 0.85

def is_flat(signal):
    return (estimateFlatness(signal) > flatness_critereon)

def top_increment(upper_limit, top, increment_size):
    return upper_limit + increment_size if (upper_limit + increment_size < top) else top 

def bottom_increment(lower_limit, bottom, increment_size):
    return lower_limit - increment_size if (lower_limit - increment_size > bottom) else bottom

def find_flat_area(signal, origin):
    
    top = len(signal)
    bottom = 0
    
    upper_limit = origin + min_window/2
    lower_limit = origin - min_window/2
    
    if (upper_limit > top):
        upper_limit = top
    if (lower_limit < bottom):
        lower_limit = bottom

    #right propagation
    while (is_flat(signal[lower_limit:top_increment(upper_limit, top, increment_size)])):
        upper_limit += increment_size
        
        #print origin, lower_limit, upper_limit, estimateFlatness(signal[lower_limit:upper_limit])
        
        if (upper_limit >= top):
            upper_limit = top
            break
        
    #left propagation
    while (is_flat(signal[bottom_increment(lower_limit, bottom, increment_size):upper_limit])):
        lower_limit -= increment_size
        
        #print origin, lower_limit, upper_limit, estimateFlatness(signal[lower_limit:upper_limit])
        
        if (lower_limit < bottom):
            lower_limit = bottom
            break
            
    return lower_limit, upper_limit
```

```python
def get_flat_areas(signal):

    origin = 0
    count = 0
    flat_areas = []
    
    while (origin < len(signal) and count != 100):
        
        l, u = find_flat_area(signal, origin)
    
        f = estimateFlatness(signal[l:u])
        
        
        #print "origin: %d/%d hit [%d, %d] with f = %g" % (origin, len(signal), l, u, f)
        
        
        if f > flatness_critereon:
            flat_areas.append([l, u])
        
        origin = u + 1
        
        
        count += 1
        
    return array(flat_areas)
```

```python
overlap = 10
silent = False
window_plotcount = 0

class window:
    
    scale = 0.5
    init_height = 5
    
    def __init__(self, signal, l, u, leftWindow = None, rightWindow = None):
        self.l = l
        self.u = u
        
        self.signal = signal.copy()
        
        self.left_neighbor = None
        self.right_neighbor = None
        
    def set_neighbors(self, left, right):
        self.right_neighbor = right
        self.left_neighbor = left
    
    def get_complement(self, _set):
        
        if len(_set) == 0:
            return [[self.l, self.u]]
        
        complement = []
        
        bottom = self.l
        top = self.u
        
        prev = bottom
        
        for unit in _set:
            
            if prev != unit[0]:
                complement.append([prev, unit[0]])
            
            prev = unit[1]

        if _set[-1][1] != top:
            complement.append([_set[-1][1], top])
            
        return complement
        
    def flatten(self):
        
        i = randint(0, 2)
        print "c =", i
        
        if i == 0:
            l = self.l
            u = (self.l + self.u)/2
        else:
            l = (self.l + self.u)/2
            u = self.u
        
        if not silent: print "flattening", l, u
        
        self.signal[l:u] = self.signal[self.l:self.u].mean() + 0.1*(-1 + 2*uniform(size=(u-l)))
    
    def get_overlap_limits(self, l, u):
        
        if (u - l < min_window):
            
            if not silent: print "need to increase window"
            
            s = int(min_window/2)
                
            if (l < s):
                l = 0
                u = min_window
                
            elif (u > N - s):
                u = N
                l = N - min_window
                
            else:     
                extra = 2*s - min_window
            
                u += s + extra 
                l -= s
        
        if not silent: print "window before overlap: ", l, u
        
        if (l < overlap):
            l = 0
            u += overlap
        elif (u > N - overlap):
            u = N
            l -= overlap
        else:
            l -= overlap
            u += overlap
        
        return l, u
    
    def setup_sub_windows(self, flat_areas):
        
        if len(flat_areas) == 0:
            return [], [self], [self]
        
        rough_areas = self.get_complement(flat_areas)
        
        for l, u in rough_areas:
            if not silent: print "rough on", l, u
        
        if len(rough_areas) == 0:
            return [self], [], [self]
        
        flat_windows  = [window(self.signal, l, u) for l, u in sorted(flat_areas , key = lambda x : x[0])]
        rough_windows = [window(self.signal, *self.get_overlap_limits(l, u)) for l, u in sorted(rough_areas, key = lambda x : x[0])]
        
        all_windows = sorted(rough_windows + flat_windows, key = lambda x : x.l)
        
        all_windows[0].set_neighbors(None, all_windows[1])
        all_windows[-1].set_neighbors(all_windows[-2], None)
        
        for i in range(1, len(all_windows) - 1):
            all_windows[i].set_neighbors(all_windows[i-1], all_windows[i+1])
        
        return flat_windows, rough_windows, all_windows
        
    
    def merge_windows(self):
        
        if len(self.sub_windows) == 0:
            return
        
        
        for w in self.sub_windows:
            self.signal[w.l:w.u] = w.signal[w.l:w.u]
        print estimateFlatness(self.signal)
        
        for i in range(len(self.sub_windows) - 1):
            
            sl = self.sub_windows[i].signal
            sr = self.sub_windows[i+1].signal
            
            rl = self.sub_windows[i].l
            ru = self.sub_windows[i].u
            ll = self.sub_windows[i+1].l
            lu = self.sub_windows[i+1].u
            
            ml = sl[ll:lu].mean()
            mr = sr[rl:ru].mean()
            
            self.sub_windows[i+1].signal[rl:ru] += ml - mr
            
        for w in self.sub_windows:
            self.signal[w.l:w.u] = w.signal[w.l:w.u]
        print estimateFlatness(self.signal)
        print "---"
    
    def plot(self, F, R):
        
        global window_plotcount
        
        k = 2
        for f, r in zip(F, R):
            h = window_plotcount*self.scale + self.init_height
            
            plot([r.l, r.u], [h, h], "r--o", markersize=5*k)
            plot([f.l, f.u], [h, h], "g--o", markersize=5*k)
            
            k += 2
            
        window_plotcount += 1
            
    def calculateWindow(self):

        if (self.left_neighbor or self.right_neighbor):
            self.flatten()
        
        flat_areas = self.l + get_flat_areas(self.signal[self.l:self.u])
        
        if not silent: print "calculating windows for", self.l, self.u, len(flat_areas)
        
        for l, u in flat_areas:
            if not silent: print "flat on", l, u
        
        flat_windows, rough_windows, self.sub_windows = self.setup_sub_windows(flat_areas)
        
        self.plot(flat_windows, rough_windows)
        
        for rough_window in rough_windows:
            rough_window.calculateWindow()

        self.merge_windows()

def findNewWindows(signal):
    w = window(signal, 0, N) 
    
    w.calculateWindow()
    
    return w

```
